```python
import seaborn as sns
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
df1=df[['PURCHASES','ONEOFF_PURCHASES','PURCHASES_FREQUENCY','PURCHASES_INSTALLMENTS_FREQUENCY','CASH_ADVANCE_TRX']]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df1= scaler.fit_transform(df1)
```

```python
from sklearn.decomposition import PCA
pca_components=PCA(n_components=2).fit_transform(df1)
pca_df = pd.DataFrame(data=pca_components ,columns=["PCA1","PCA2"])
```

```python
pca_df.head()
```

```python
from sklearn.cluster import KMeans

wss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init="k-means++",max_iter=300,random_state=0)
    kmeans.fit(pca_df)
    wss.append(kmeans.inertia_)
    
    
```

```python
plt.plot(range(1,11),wss,marker='o')
plt.title('elbow method for find K value')
plt.xlabel('number of cluster')
plt.ylabel('wss')
plt.show()
```

```python
K=3
kmeans=KMeans(n_clusters=K,init="k-means++")
kmeans.fit(pca_df)
labels=kmeans.labels_
print(labels)
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

labels = kmeans.labels_

print("Silhouette Score:", silhouette_score(pca_df, labels))
print("Calinski-Harabasz Score:", calinski_harabasz_score(pca_df, labels))
print("Davies-Bouldin Score:", davies_bouldin_score(pca_df, labels))
```

```python
pca_df_pd=pd.concat([pca_df,pd.DataFrame({'cluster':labels})],axis=1)
```

```python
plt.Figure(figsize=(10,10))
gv=sns.scatterplot(x='PCA1',y='PCA2',hue='cluster',data=pca_df_pd,palette=['red','green','blue'])
plt.title('clustering into 3 groups')
plt.show()
```
